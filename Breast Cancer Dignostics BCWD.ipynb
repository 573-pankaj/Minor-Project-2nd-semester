```python
models=[]

models.append(('KNN', KNeighborsClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
```

```python
# evaluate each model

results =[]
names=[]
for name , model in models:
    kfold=KFold(n_splits=10, random_state=None)
    cv_results= cross_val_score(model, X_train, y_train, cv=kfold,scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    
    msg= '%s:, %f, (%f)' % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
from contextlib import contextmanager
import time
@contextmanager
def timer():
    """Helper for measuring runtime"""

    time0 =  time.perf_counter()
    #time.sleep(1)
    yield
    print('[elapsed time: %.18f s]' % (time.perf_counter() - time0))
```

# SVM

```python
# make predictions on validation datasets
with timer() as t:
    SVM = SVC()
    SVM.fit(X_train, y_train)
    predictions= SVM.predict(X_test)
    print(("training score : {}").format(SVM.score(X_train, y_train)))
    print(("testing score : {}").format(accuracy_score(y_test, predictions)))
    print(classification_report(y_test, predictions))
    print(confusion_matrix(y_test, predictions))
    
```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr(method='spearman')
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.60) # 0.65
len(set(corr_features))
```
